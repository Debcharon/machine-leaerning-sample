```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.linspace(-10, 10, 200)
y1 = x**2
y2 = x

# 创建图形和子图
fig, (ax1, ax2) = plt.subplots(1, 2)

# 第一个子图
ax1.plot(x, y1)
ax1.set_title('y = x^2')

# 第二个子图
ax2.plot(x, y2)
ax2.set_title('y = x')

# 显示图形
plt.show()
```

```python
# 创建两个子图
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# 在第一个子图中
axs[0].plot(x, y1, color="red", label="y = x^2")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].legend()

# 在第二个子图
axs[1].plot(x, y2, color="blue", label="y = x")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].legend()

# 显示图形
plt.show()
```
